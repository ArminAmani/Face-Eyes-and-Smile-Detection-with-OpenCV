```python
import cv2 as cv
import os
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Face + Eye + Smile Detection

```python
Face_Cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_frontalface_default.xml")
Eye_Cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_eye.xml")
Smile_Cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_smile.xml")
```

* **These are like little robots that know how to find faces, eyes, and smiles.**
* **You donâ€™t need to download anything â€” OpenCV already has them built-in!**

```python
def face_detection(bw_image, orig_image):
    faces = Face_Cascade.detectMultiScale(bw_image, scaleFactor=1.1, minNeighbors=5, minSize=(80, 80))

    for (x, y, w, h) in faces:
        cv.rectangle(orig_image, (x, y), (x + w, y + h), (255, 0, 0), 3)
        cv.putText(orig_image, 'Face', (x, y - 10), cv.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

        roi_gray = bw_image[y:y+h, x:x+w]
        roi_color = orig_image[y:y+h, x:x+w]


        # eye detection
        eyes = Eye_Cascade.detectMultiScale(roi_gray, scaleFactor=1.05, minNeighbors=10)
        for (ex, ey, ew, eh) in eyes:
            cv.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 1)

        # smile detection ====> Smile is very hard with Haar â€” needs good lighting + open mouth
        smiles = Smile_Cascade.detectMultiScale(roi_gray, scaleFactor=1.4, minNeighbors=20, minSize=(40, 40))
        for (sx, sy, sw, sh) in smiles:
            cv.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (0, 0, 255), 1)
            cv.putText(roi_color, "Smile", (sx, sy-10), cv.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1)

    return orig_image
```

```python
# Screenshot function (fixed path)
def make_screenshot(img, counter):
    img_dir = "face_detection_screenshots"
    os.makedirs(img_dir, exist_ok=True)
    path = os.path.join(img_dir, f"screenshot_{counter:04d}.jpg")
    cv.imwrite(path, img)
    print(f"Saved {path}")
```

```python
cap = cv.VideoCapture(0)
screenshot_counter = 0

print("Press 's' to save screenshot, 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    result = face_detection(gray, frame.copy())

    cv.imshow('Face + Eye + Smile Detection', result)

    key = cv.waitKey(1) & 0xFF
    if key == ord('s'):
        make_screenshot(result, screenshot_counter)
        screenshot_counter += 1
    elif key == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
